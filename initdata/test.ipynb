```python
import pandas as pd
```

```python
# 老师
teacher = pd.read_csv('teachers.csv',header = 0)
print(teacher)
teacher['teacher_type'] = teacher['teacher_type'].replace([0,1],['教师','主任'])
print(teacher.columns)

with open("teacher.txt","w",encoding='utf-8') as tea:
    for row in teacher.iterrows():
        tea.write("INSERT INTO teacher VALUES ({0}, '{1}', '{2}', '{3}', '{4}', {5});".format(row[1][0],row[1][1],row[1][2],row[1][3],row[1][4],row[1][5])+'\n')
```

```python
classes = pd.read_csv('classes.csv', header=0, )
with open("classes.txt","w",encoding='utf-8') as cla:
    for row in classes.iterrows():
        cla.write("INSERT INTO class VALUES ({0}, {1}, '{2}');".format(row[1][0],row[1][1],row[1][2])+'\n')
```

```python
# student
student = pd.read_csv('students.csv', header=0)
with open("student.txt","w",encoding='utf-8') as stu:
    for row in student.iterrows():
        stu.write("INSERT INTO student VALUES ({0}, {1}, '{2}', '{3}', to_date('2018-09-01','YYYY-MM-DD'),'计算机科学与技术', 123456 );".format(row[1][0],row[1][1],row[1][2], row[1][3])+'\n')

```

```python
# courses
courses = pd.read_csv('courses.csv', header=0)
courses['is_compulsory'] = courses['is_compulsory'].replace([0, 1],['专业选修课', '专业必修课'])
with open("courses.txt","w",encoding='utf-8') as stu:
    for row in courses.iterrows():
        stu.write("INSERT INTO course VALUES ({0}, '{1}', '{2}', '{3}', {4}, {5});".format(row[1][0],row[1][1], row[1][3], row[1][4], row[1][2], row[1][5])+'\n')
```

```python
teacher_group = pd.read_csv('teacher_group.csv', header=0)
with open("teacher_group.txt","w",encoding='utf-8') as stu:
    for row in teacher_group.iterrows():
        stu.write("INSERT INTO teacher_group VALUES ({0}, '{1}');".format(row[1][0],row[1][1])+'\n')
```

```python
def split_teacher( teacher_group, classes):
    dic = {}
    for i in teacher_group:
        if i[0] not in dic.keys():
            dic[i[0]] = set([i[1]])
        else:
            dic[i[0]].add(i[1])
    print(dic)

    for key in dic:
        tem_list = list(dic[key])

        if len(tem_list) == 1:
            for i in range(2):
                for j in range(3):
                    print("新课程id{0}{1}, 任课老师id{2}, 班别id{3}, 课程原id{4}".format(key,i,tem_list[0], classes[3*i+j][0], key))
        elif len(tem_list) == 2:
            for i in range(2):
                for j in range(3):
                    print("新课程id{0}{1}, 任课老师id{2}, 班别id{3}, 课程原id{4}".format(key,i,tem_list[i], classes[3*i+j][0], key))
        elif len(tem_list) == 3:
            for i in range(3):
                for j in range(2):
                    print("新课程id{0}{1}, 任课老师id{2}, 班别id{3}, 课程原id{4}".format(key,i,tem_list[i], classes[2*i+j][0], key))
        elif len(tem_list) == 4:
            for i in range(2):
                for j in range(2):
                    print("新课程id{0}{1}, 任课老师id{2}, 班别id{3}, 课程原id{4}".format(key,i,tem_list[i], classes[2*i+j][0], key))
            print("新课程id{0}{1}, 任课老师id{2}, 班别id{3}, 课程原id{4}".format(key,2,tem_list[2], classes[4][0], key))
            print("新课程id{0}{1}, 任课老师id{2}, 班别id{3}, 课程原id{4}".format(key,3,tem_list[3], classes[5][0], key))
        print(key, dic[key],len(dic[key]))


split_teacher(teacher_group.values, classes.values)
```
